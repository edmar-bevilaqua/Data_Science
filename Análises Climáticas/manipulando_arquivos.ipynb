## Declarando a função para deletar arquivos, bem como log de sucesso ou fracasso
---

```python
def delete_file(filename):
    try:
        os.remove(filename)
        print("File " + filename + " deleted successfully.")
    except IOError:
        print("Error: could not delete file " + filename)
```

## Excluindo os arquivos que não possuem a string especificada.
---

```python
for i in arquivos_com_caminho:
    if not "INMET_S_RS" in i:
        delete_file(i)
```
