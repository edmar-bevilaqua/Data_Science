```python
m1 = 'CCcCCccCCCccCcCccCcCcCCCcCCcccCCcCcCcCcccCCcCcccCc'
m2 = 'CCCCCccCccCcCCCCccCccccCccCccCCcCccCcCcCCcCccCccCc'
m3 = 'CccCCccCcCCCCCCCCCCcccCccCCCCCCccCCCcccCCCcCCcccCC'
m4 = 'cCCccCCccCCccCCccccCcCcCcCcCcCcCCCCccccCCCcCCcCCCC'
m5 = 'CCCcCcCcCcCCCcCCcCcCCccCcCCcccCccCCcCcCcCcCcccccCc'
```

```python
eventos = {'m1': list(m1), 
            'm2': list(m2), 
            'm3': list(m3), 
            'm4': list(m4), 
            'm5': list(m5)}
eventos
```

```python
moedas = pd.DataFrame(eventos)
moedas
```

```python
df = pd.DataFrame(data = ['Cara', 'Coroa'], 
                    index = ['c', 'C'], 
                    columns = ['Faces'])
df
```

```python
for item in moedas:
  df = pd.concat([df, moedas[item].value_counts()], axis= 1)
df
```

```python
import random
```

```python
v1 = []
for _ in range(50):
  v1.append(random.randint(1, 1000))
v2 = []
for _ in range(50):
  v2.append(random.randint(1, 1000))
v3 = []
for _ in range(50):
  v3.append(random.randint(1, 1000))
v4 = []
for _ in range(50):
  v4.append(random.randint(1, 1000))
v5 = []
for _ in range(50):
  v5.append(random.randint(1, 1000))
```

```python
variaveis = pd.DataFrame(data={'v1' : v1,
                               'v2' : v2,
                               'v3' : v3,
                               'v4' : v4,
                               'v5' : v5})
variaveis
```

```python
for item in variaveis:
  variaveis['imp_par_' + str(item)] = variaveis[item].apply(lambda x: 'Par' if x % 2 == 0 else 'Impar')
```

```python
variaveis
```

```python
variaveis.drop(['v1','v2','v3','v4','v5'], axis=1, inplace=True)
variaveis
```

```python
contadores = pd.DataFrame(
    data= ['Impar', 'Par'],
    columns= ['Classe'],
    index= ['Impar', 'Par']
)
contadores
```

```python
for item in variaveis:
  contadores = pd.concat([contadores, variaveis[item].value_counts()], axis= 1)
  print(contadores)
```

```python
contadores.drop(['Classe'], axis= 1, inplace= True)
```
