## Criando uma Series

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
```

```python
s
```

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s.fillna(method = 'ffill')
s.ffill().mean()
```

```python
s.bfill().mean()
```

```python
s.fillna(s.mean()).mean()
```

```python
s.mean()
```

```python
s.fillna(method = 'ffill', limit = 1)
```

```python
s
```

```python
s1 = s.fillna(method= 'ffill', limit= 1)
s1.fillna(method='bfill', limit= 1)
```

```python
atletas = pd.DataFrame([['Marcos', 9.62], ['Pedro', None], ['João', 9.69], 
                        ['Beto', 9.72], ['Sandro', None], ['Denis', 9.69], 
                        ['Ary', None], ['Carlos', 9.74]], 
                        columns = ['Corredor', 'Melhor Tempo'])
atletas
```

```python
atletas.fillna(atletas.mean(), inplace= True)
atletas
```

```python
atletas.fillna(atletas["Melhor Tempo"].mean(), inplace = True)
atletas
```

```python
selection = atletas['Melhor Tempo'].notnull()
selection
```

```python
atletas = atletas[selection]
atletas
```

```python
atletas = atletas[atletas['Melhor Tempo'].notnull()]
atletas
```
