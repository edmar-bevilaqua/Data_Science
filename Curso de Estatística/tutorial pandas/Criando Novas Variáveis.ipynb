```python
import pandas as pd
```

```python
df = pd.read_csv('/content/aluguel_residencial (1).csv', sep= ';', header= 0)
df.head(10)
```

```python
df['Valor Bruto'] = df['Valor'] + df['Condominio'] + df['IPTU']
df.head(10)
```

```python
df['Valor m2'] = (df['Valor'] / df['Area']).round(2)
df.head(10)
```

```python
df['Valor Bruto m2'] = (df['Valor Bruto'] / df['Area']).round(2)
df.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
df['Tipo Agregado'] = df['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
df.head(10)
```

```python
import pandas as pd
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas'])
```

```python
alunos
```

```python
alunos['Faixa Etária'] = alunos['Idade'] \
    .apply(lambda x: 'Menor que 20 anos' if x < 20 
        else ('Entre 20 e 40 anos' if (x >= 20 and x <= 40) 
            else 'Maior que 40 anos'))
alunos
```
