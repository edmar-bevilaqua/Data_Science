### Fazendo as primeiras visualizações

```python
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df['celsius'] = df['temperatura'] - 273
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', ls = 'dotted')

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
import datetime
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_xlim(left = datetime.datetime(2014,1,1), right = datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', marker = 'o')

eixo.set_xlim(left = datetime.datetime(2014,1,1), right = datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', lw = 0.7)

eixo.set_xlim(left = datetime.datetime(2014,1,1), right = datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
eixo.grid(visible = True)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo_2 = fig.add_axes([0.75, 0.7, 0.25, 0.27])

eixo.set_xlim(left = datetime.datetime(2014, 1, 1), right = datetime.datetime(2015, 1, 1))
eixo.grid(visible = True)
eixo.plot(df['data'], df['temperatura'], color = 'blue', lw = 0.7)
eixo.set_title(label = 'Temperatura x Data', fontsize = 20, pad = 10)
eixo.legend(['Temperatura'], loc = 'upper left')

eixo_2.plot(df['data'], df['temperatura'], color = 'green', lw = 0.7)
eixo_2.set_title(label = 'Temperatura x Data', fontsize = 14)
eixo_2.legend(['Temperatura'], loc = 'upper left')
eixo_2.legend(['Temperatura'], loc = 'upper left')
eixo_2.grid(True)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo_2 = fig.add_axes([0.75, 0.7, 0.25, 0.27])

eixo.set_xlim(left = datetime.datetime(2014, 5, 1), right = datetime.datetime(2014, 5, 30))
eixo.grid(visible = True)
eixo.plot(df['data'], df['celsius'], color = 'blue', lw = 0.7)
eixo.set_title(label = 'Temperatura x Data', fontsize = 20, pad = 10)
eixo.legend(['Temperatura'], loc = 'upper left')
eixo.set_ylim(bottom= 0, top = 40)

eixo_2.set_xlim(left = datetime.datetime(2014, 1, 1), right = datetime.datetime(2014, 12, 31))
eixo_2.plot(df['data'], df['celsius'], color = 'green', lw = 0.7)
eixo_2.set_title(label = 'Temperatura x Data', fontsize = 14)
eixo_2.legend(['Temperatura'], loc = 'upper left')
eixo_2.legend(['Temperatura'], loc = 'upper left')
eixo_2.grid(True)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo_2 = fig.add_axes([0.75, 0.7, 0.25, 0.27])

eixo.set_xlim(left = datetime.datetime(2014, 5, 1), right = datetime.datetime(2014, 5, 30))
eixo.grid(visible = True)
eixo.plot(df['data'], df['celsius'], color = 'green', lw = 1.2)
eixo.set_title(label = 'Temperatura x Data', fontsize = 20, pad = 10)
eixo.legend(['Temperatura'], loc = 'upper left')
eixo.set_ylim(bottom= 0, top = 40)

left_blue = df['data'] < datetime.datetime(2014, 5, 1)
right_blue = df['data'] > datetime.datetime(2014, 5, 30)

eixo_2.set_xlim(left = datetime.datetime(2014, 1, 1), right = datetime.datetime(2014, 12, 31))
eixo_2.plot(df['data'], df['celsius'], color = 'green', lw = 0.7)
eixo_2.plot(df[left_blue]['data'], df[left_blue]['celsius'], color = 'blue')
eixo_2.plot(df[right_blue]['data'], df[right_blue]['celsius'], color = 'blue')
eixo_2.set_title(label = 'Temperatura x Data', fontsize = 14)
eixo_2.legend(['Temperatura'], loc = 'upper left')
eixo_2.legend(['Temperatura'], loc = 'upper left')
eixo_2.grid(True)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['celsius'], color = 'green', lw = 1)
eixo.set_title('Temperatura x Tempo', fontsize = 25, pad= 10)
eixo.set_xlabel('Data', fontsize = 18)
eixo.set_ylabel('Temperatura (°C)', fontsize = 18)
eixo.legend(['Temperatura'])
eixo.grid(visible=True)
eixo.axhline(y = max(df['celsius']), color = 'k', ls='--')
eixo.axhline(y = min(df['celsius']), color = 'k', ls='--')
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['celsius'], color = 'green', lw = 1)
eixo.set_title('Temperatura x Tempo', fontsize = 25, pad= 10)
eixo.set_xlabel('Data', fontsize = 18)
eixo.set_ylabel('Temperatura (°C)', fontsize = 18)
eixo.legend(['Temperatura'], loc = 'lower left')
eixo.grid(visible=True)
eixo.axhline(y = max(df['celsius']), color = 'k', ls='--')
eixo.axhline(y = min(df['celsius']), color = 'k', ls='--')

x1 = df['data'][df['celsius'].idxmax()]
y1 = max(df['celsius'])

eixo.annotate('Máx', xy = (x1, y1))
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['celsius'], color = 'green', lw = 1)
eixo.set_title('Temperatura x Tempo', fontsize = 25, pad= 10)
eixo.set_xlabel('Data', fontsize = 18)
eixo.set_ylabel('Temperatura (°C)', fontsize = 18)
eixo.legend(['Temperatura'], loc = 'lower left')
eixo.grid(visible=True)
eixo.axhline(y = max(df['celsius']), color = 'k', ls='--')
eixo.axhline(y = min(df['celsius']), color = 'k', ls='--')

x1 = df['data'][df['celsius'].idxmax()]
y1 = max(df['celsius'])

x2 = df['data'][df['celsius'].idxmax() - 7000]
y2 = max(df['celsius']) - 3

eixo.annotate('Máximo', xy = (x1, y1), fontsize = 15, xytext= (x2, y2), arrowprops=dict(facecolor = 'k'))

x3 = df['data'][df['celsius'].idxmin()]
y3 = min(df['celsius'])

x4 = df['data'][df['celsius'].idxmin() - 7000]
y4 = min(df['celsius']) + 3
eixo.annotate('Mínimo', xy = (x3, y3), fontsize = 15, xytext= (x4, y4), arrowprops=dict(facecolor = 'k'))
```
