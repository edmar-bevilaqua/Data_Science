## DBSCAN - Demo 3

* Use of the DBSCAN algorithm with different parameters.
* Synthetic database
* How to choose the best `epsilon` and `minpoints` parameters

```python
import time
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs

%matplotlib inline

sns.set(style='darkgrid', context='talk', rc={'figure.figsize': (14, 8)})
```

## Creating a Synthetic Dataset

```python
X, y = make_blobs(n_samples=1000, centers=7, n_features=2,
                  random_state=33)

df = pd.DataFrame(X, columns=['x', 'y'])
df['label'] = y

df.head()
```

```python
_ = sns.scatterplot(x='x', y='y', data=df, hue='label', palette='rainbow', legend=False)
```

```python
def train_and_plot(data, eps=0.5, min_samples=5):
    
    # Copiamos os dados e removemos a coluna label 
    data_copy = data.copy()
    data_copy.drop(columns='label', inplace=True)
    
    model = DBSCAN(eps=eps, min_samples=min_samples)
    model.fit(data_copy)
    
    data_copy['cluster_labels'] = model.labels_
    _ = sns.scatterplot(x='x', y='y', data=data_copy, hue='cluster_labels', palette='rainbow', legend=False)
    
    return model.labels_
```

### How to choose DBSCAN parameters

One of the ways you can choose DBSCAN's parameters is by using what we call a `K-dist plot`. Let's see how it works.

To generate the K-dist plot, we will use KNN

```python
from sklearn.neighbors import NearestNeighbors
```

We define a `n_neighbors` value according to the `min_samples` value we want in DBSCAN.

```python
n_neighbors = 5
knn = NearestNeighbors(n_neighbors=n_neighbors)
neighbors = knn.fit(X)
```

We obtain the distances and the corresponding indices

```python
distances, indices = neighbors.kneighbors(X)
```

```python
distances
```

We sort the distances from the K-th nearest neighbor (in this case, the 5th) in ascending order.

```python
indices
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:, n_neighbors-1]
```

We make the plot and look for where there is a sharp increase.

```python
plt.plot(distances)
```

We choose the value that seems best. In this case, we can use a value between `0.5` and `0.8`.

```python
eps = 0.6
min_samples = 5

labels = train_and_plot(df, eps, min_samples)
```

```python
np.unique(labels, return_counts=True)
```

## Running on Different Datasets

```python
def do_kdist_plot(X, n_neighbors=5):
    knn = NearestNeighbors(n_neighbors=n_neighbors)
    neighbors = knn.fit(X)
    
    distances, _ = neighbors.kneighbors(X)
    distances = np.sort(distances, axis=0)
    distances = distances[:, n_neighbors-1]
    
    plt.plot(distances)
```

```python
transformation = [[0.3, -0.3], [-1.2, 0.2]]

X, y = make_blobs(n_samples=1000, centers=7, n_features=2,
                  random_state=170)

df_transformed = pd.DataFrame(np.dot(X, transformation), columns=['x', 'y'])
df_transformed['label'] = y

df_transformed.head()
```

```python
_ = sns.scatterplot(x='x', y='y', data=df_transformed, hue='label', palette='rainbow', legend=False)
```

### Example using 5 neighbors:

```python
do_kdist_plot(X, n_neighbors=5)
```

```python
labels = train_and_plot(df_transformed, eps=0.4, min_samples=5)
```

### Example using 10 neighbors:

```python
do_kdist_plot(X, n_neighbors=10)
```

```python
labels = train_and_plot(df_transformed, eps=0.5, min_samples=10)
```

```python
X, y = make_moons(300, noise=.05, random_state=33)

df_moon = pd.DataFrame(X, columns=['x', 'y'])
df_moon['label'] = y

df_moon.head()
```

```python
_ = sns.scatterplot(x='x', y='y', data=df_moon, hue='label', palette='rainbow', legend=False)
```

```python
do_kdist_plot(X, n_neighbors=8)
```

```python
labels = train_and_plot(df_moon, eps=0.15, min_samples=10)
```

# Conclusion:
We can arrive at the best value for `eps` and `min_samples` by using the KNN algorithm to obtain the distance to the k-th member, and this distance is analyzed on the graph, allowing us to choose the best value for `eps`.
