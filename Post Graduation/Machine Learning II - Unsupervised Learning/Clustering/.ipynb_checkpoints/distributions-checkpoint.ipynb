## Differences in distributions with different means and standard deviations

```python
np.random.seed(41)

mu = [0, -2, -1, 5]
sigma = [0.1, 0.2, 0.4, 0.8]
size = 1000

df = pd.DataFrame(columns=["value", "mu"])
for i in range(len(mu)):
    data = np.random.normal(mu[i], sigma[i], size)
    data = pd.DataFrame(data, columns=["value"])
    data["mu"] = mu[i]
    df = pd.concat([data, df])

display(df)
```

```python
ax = sns.displot(data=df, x='value', hue='mu', kind='kde', fill=True, palette=sns.color_palette('bright')[:4], aspect=2, height=8)
```

## Different types of distribution

There are several types of distribution and each type has its different applications. Some distributions apply to discrete variables while others apply to continuous variables. Here we will see, graphically, how each type behaves.

### Uniform
All the values in the interval are equally likely.

#### Examples
* When we throw a dice, the chance of getting 1, 2, 3, 4, 5 or 6 is equally likely. 
* When we pick a card from a deck (without repeated cards), it is equally likely that the card is a King of Spades, a Queen of Hearts, or any other card.

```python
data = uniform.rvs(size=10000, loc=0, random_state=random_seed)
ax = sns.displot(data, kind='kde', fill=False, aspect=2, height=6, color='b')
```

### Poisson

Probability distribution related to the number of times a given event occurred in a given time interval (or space).

\begin{align}
\mathbb{P}(X=x)=e^{-\lambda}\frac{\lambda^{x}}{x!}
\end{align}

$e$ = Euler's Number

$\lambda$ = Average number of events in a given interval.

#### Examples
* What is the probability that a salesman sells 3 apartments in 1 week, when the average number of apartments is 10 per month?
* What is the probability of 4 babies being born in one hour in a hospital, when the average number of births is 1.8 babies per hour?

```python
data = poisson.rvs(size=10000, mu=3, random_state=random_seed)
ax = sns.displot(data, kind='kde', fill=False, aspect=2, height=6, color='g')
```

### Binomial

Probability distribution that describes the number of successes in a sequence of $N$ independent experiments, where each experiment is binary (e.g. yes or no, heads or tails).

\begin{align}
\mathbb{P}(X=x)=\left(\begin{array}{c}n\\x\end{array}\right) p^x(1-p)^{n-x}
\end{align}

\begin{align}
\left(\begin{array}{c}n\\x\end{array}\right)=\frac{n!}{x!(n-x)!}
\end{align}

#### Examples
* Imagine we throw a dice 10 times. What is the probability of getting the number "6" five times?
* Imagine we have a “faulty” coin, where we get tails with probability 0.3. What is the probability of getting heads after tossing? 

```python
data = binom.rvs(n=5, p=0.8, size=100, random_state=random_seed)
ax = sns.displot(data, kind='kde', fill=False, aspect=2, height=6, color='r')
```

### Exponential
Used to describe the interval or space in which something occurs. It has a direct link with the Poisson distribution:
* Poisson: estimates the number of events in an interval
* Exponential: analyzes the opposite - an interval or space in which an event occurs.

#### Examples
* The average service time at a bank teller is 7 minutes. What is the probability that a customer will wait longer than 10 minutes to be served, given that the average time is described by an exponential distribution?
* On average, a student completes 6 math exercises in 1 hour. What is the probability of a student finishing an exercise between 5 and 10 minutes?

```python
data = expon.rvs(scale=1, loc=0, size=1000, random_state=random_seed)
ax = sns.displot(data, kind='kde', fill=False, aspect=2, height=6, color='y')
```

## Central Limit Theorem

```python
sns.set(style='darkgrid', rc={'figure.figsize': (14, 18)})
```

```python
def plot_distributions_comparison(distribution='uniform', num_samples=1000):

    plot_idx = 0
    max_iter = 101
    
    f, axes = plt.subplots(6, 2)
    plt.subplots_adjust(bottom=0.0, top=1.2)

    f.suptitle('Distribution: ' + distribution, y=1.25)
    
    for i in range(0, max_iter, 20):
        if i == 0:
            num_iter = 2
        else:
            num_iter = i

        data = []
        data_mean = []

        for j in range(num_iter):
            
            if distribution == 'uniform':
                value = uniform.rvs(size=num_samples,loc=-2)
            elif distribution == 'poisson':
                value = poisson.rvs(size=num_samples,loc=-2, mu=1)
            elif distribution == 'exponential':
                value = expon.rvs(scale=1, loc=-2, size=num_samples)
            elif distribution == 'binomial':
                value = binom.rvs(n=10, p=0.8, size=num_samples)         
            
            data_mean.append(np.mean(value))
            data.extend(value)

        title = 'Number of samples: ' + str(num_iter)
        ax = sns.histplot(data, kde=True, ax=axes[plot_idx][0])
        ax.set_title(title)

        ax = sns.histplot(data_mean, kde=True, ax=axes[plot_idx][1])
        ax.set_title(title + ' - mean')

        plot_idx += 1
```

### Uniform

```python
plot_distributions_comparison('uniform')
```

### Poisson

```python
plot_distributions_comparison('poisson')
```

### Binomial

```python
plot_distributions_comparison('binomial')
```

### Exponential

```python
plot_distributions_comparison('exponential')
```
