### DBSCAN - Demo 2

* Use of the DBSCAN algorithm with different parameters.
* Synthetic database with different formats

```python
import time
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.datasets import make_circles
from sklearn.datasets import make_s_curve

%matplotlib inline

sns.set(style='darkgrid', context='talk', rc={'figure.figsize': (14, 8)})
```

### Creating Function to Test DBSCAN

```python
def train_and_plot(data, eps=0.5, min_samples=5):
    
    # Copiamos os dados e removemos a coluna label 
    data_copy = data.copy()
    data_copy.drop(columns='label', inplace=True)
    
    model = DBSCAN(eps=eps, min_samples=min_samples)
    model.fit(data_copy)
    
    data_copy['cluster_labels'] = model.labels_
    _ = sns.scatterplot(x='x', y='y', data=data_copy, hue='cluster_labels', palette='rainbow', legend=False)
    
    return model.labels_
```

### Creating Synthetic Dataset with Different Shapes

```python
X, y = make_circles(n_samples=1000, factor=0.3, noise=0.1)

df = pd.DataFrame(X, columns=['x', 'y'])
df['label'] = y

df.head()
```

```python
_ = sns.scatterplot(x='x', y='y', data=df, hue='label', palette='rainbow', legend=False)
```

## Testing DBSCAN

```python
eps = 0.15
min_samples = 5

labels = train_and_plot(df, eps, min_samples)
```

```python
np.unique(labels, return_counts=True)
```

### Creating Another Synthetic Dataset with a Different Shape

```python
transformation = [[1.3, 2.3], [1.2, 5.2]]

X, y = make_blobs(n_samples=1000, centers=7, n_features=2,
                  random_state=170)

df_transformed = pd.DataFrame(np.dot(X, transformation), columns=['x', 'y'])
df_transformed['label'] = y

df_transformed.head()
```

```python
_ = sns.scatterplot(x='x', y='y', data=df_transformed, hue='label', palette='rainbow', legend=False)
```

```python
labels = train_and_plot(df_transformed, eps=1.5, min_samples=5)
```

```python
np.unique(labels, return_counts=True)
```

# Conclusion:
We can see that DBSCAN copes well with problems where we have clusters that surround others, as in the case where we had a cluster that was a circle with another cluster inside. It is also possible to verify that DBSCAN has difficulties dealing with clusters that are very close to each other, causing the algorithm to often consider the 2 clusters as 1.
